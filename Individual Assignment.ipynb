```python
import numpy as np
import pandas as pd
imdb = pd.read_csv ('https://raw.githubusercontent.com/scalabretta/GroupProject--IMDB/master/Dataset/IMDB-database-update.csv',
                    encoding = 'latin1')
```

```python
# 1. How many rows and columns reflected on dataset file IMDB-database-update.csv
imdb.shape
```

```python
# 2. What are the top 3 movies with highest nett profit in 2013? 
imdb_2013 = (imdb[imdb.title_year == 2013])
imdb_2013.sort_values(by=['net_USD'], ascending=False).head(3)
```

```python
# 3. What are the top 3 movies with highest gross profit in 2009?
imdb_2009 = (imdb[imdb.title_year == 2009])
imdb_2009.sort_values(by=['gross_USD'], ascending=False).head(3)
```

```python
# 4. How many non-english movies are in the dataset? 
non_english = (imdb[imdb.language != 'English'])
print(len(non_english))
```

```python
# 5. What are top 3 languages in the dataset aside English (based on number of movies)? 
non_english['language'].value_counts ().head(3)
```

```python
# 6. What are the top 3 countries with the highest nett profit in all dataset times? 
imdb.sort_values(by=['net_USD'], ascending=False).head(3)
```

```python
# 7. What are the average budget of all movies in dataset in 2013?
print("Rata-rata pengeluaran 2013 = ", round(imdb_2013[['gross_USD']].mean(),2))
```

```python
# 8. What are the average budget of all movies in dataset in 2014? 
imdb_2014 = (imdb[imdb.title_year == 2014])
print("Rata-rata Budget 2014 = ", round(imdb_2014[['budget_USD']].mean(),2))
```

```python
# 9. Which one is higher, average budget of all movies in dataset in 2014 or 2013? 
budget_2013 = round(imdb_2013[['budget_USD']].mean(),2)
budget_2014 = round(imdb_2014[['budget_USD']].mean(),2)
budget = [budget_2013.values[0], budget_2014[0]]
print('budget 2013 =',budget_2013[0], "\n"
      'budget 2014 =',budget_2014[0], "\n"
      'average budget of all movies in dataset (2013) =',max(budget))
```
