#### Convert Dict to DataFrame

```python
code  = ['US', 'AUS', 'JPN', 'IN', 'RUS', 'MOR', 'EG']
names = ['United States', 'Australia', 'Japan', 'India','Russia','Morocco', 'Egypt']
dr    = [True, False, False, False, True, True, True]
cpc   = [809, 731, 588, 18, 200, 70, 45]
         
# Create dictionary my_dict with three key:value pairs: my_dict
my_dict = {'ctrcode':code, 'country':names, 'drives_right':dr,'cars_per_cap':cpc}
cars = pd.DataFrame(my_dict)

print(type(cars))
print(cars)
```

#### Pandas loc and iloc

```python
cars_id = cars.set_index("ctrcode") 
print(cars_id.shape) 
cars_id
```

```python
cars_id['country'], cars_id[['country']] 
cars_id[0:3]
```

```python
# row & column access
cars_id.loc[['AUS', 'EG']] 
```

```python
 # row access
cars_id.loc[['RUS', 'MOR'], ['country', 'drives_right']]
```

#### Where Clause and Sort
* filtering with columns rather than index

```python
import pandas as pd
print(cars_id[cars_id.drives_right == True]) 
print(cars_id[(cars_id.cars_per_cap > 100) & (Data_IMDB_id.country == "Australia") ])
```
